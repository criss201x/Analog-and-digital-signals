2) Digitalization

You have to create a function that receives the analog value (aval), the power source of A2D converter (vcc), and the number of bits of the A2D converter (n). 
The function should return the digital value equivalent to the given analog value.
Assign amplitude to the signal S by using S2 = A*sin(w*t).
Plot S2 with A in range [5,10,15,20]

```python
ranges = chain(range(20), range(5,10,15))
w = 0.1
plt.figure(figsize=(30, 15))
for A in ranges:
    S2 = A*np.sin(w*t)    
    plt.title('Sine wave in range [5,10,15,20]')
    plt.xlabel('Time (t)')
    plt.ylabel('S = sin(time)')
    plt.plot(t, S2, label='A*sin(w*t)')
```
