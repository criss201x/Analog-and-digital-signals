#### Necessary tools for the workshop

```python
import matplotlib.pylab as plt
import matplotlib.pyplot as subplt
import numpy as np
plt.style.use('seaborn-poster')
%matplotlib inline
```

#### Definition of the one-dimensional array where (t) represents the time

```python
t = np.arange(start=1, stop=501, step=1)
print(t)
```

#### We now define a sinusoidal signal using an arbitrary frequency ($\omega$) at: A $ sin(\omega )  $

```python
#Amplitude of the sine wave is sine of a variable like time
w = 0.1
amplitude = np.sin(t*w)
plt.figure(figsize=(30, 15))
plt.title('Sine wave')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude = sin(time)')
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.plot(t, amplitude)
```

- Sine waves represent periodic oscillations.
- The amplitude of the sine wave at any point in Y is proportional to the sine of a variable.
- The frequency of the sine wave is given by number of cycles per second.
- A denotes amplitude of a sine wave.
